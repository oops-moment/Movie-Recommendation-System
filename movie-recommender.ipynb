```python
import ast 
```

```python
def convert(obj):
    Listt=[]
    for i in ast.literal_eval(obj): 
        Listt.append(i['name'])
    return Listt
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
# we need name for this as well only top3 but
def convert_top3(obj):
    Listt=[]
    counter=0
    for i in ast.literal_eval(obj): 
        Listt.append(i['name'])
        counter+=1
        if(counter==3):
            break
    return Listt
```

```python
movies['cast'] = movies['cast'].apply(convert_top3)
```

```python
def directordo(obj):
    Listt=[]
    for i in ast.literal_eval(obj): 
        if(i['job']=='Director'):
            Listt.append(i['name'])
    return Listt
```

```python
movies['crew'] = movies['crew'].apply(directordo)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english')
# we will use cv 
vectors=cv.fit_transform(new_df['tags']).toarray()


```

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()



```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  #convert to strig
```

```python
ps.stem("loving")
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# w want har movie distanct jitna jyad utna difference

# more high dimension eucldean not relaiable hence use cosine distance

from sklearn.metrics.pairwise import cosine_similarity

```

```python
similiarity=cosine_similarity(vectors)
```

```python
# list(enumerate(similiarity[0]))
# sorted(list(enumerate(similiarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similiarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
    return

# we can take index as awell soer krn amax same but index bhui chaiye sath
```
